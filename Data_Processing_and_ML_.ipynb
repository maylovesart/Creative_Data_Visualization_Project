```python
# Data Processing and ML for Creative Visualization

# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Generate sample data
def generate_data():
    countries = ['USA', 'China', 'India', 'Brazil', 'Russia', 'Japan', 'Germany', 'UK', 'France', 'Italy']
    data = pd.DataFrame({
        'country': countries,
        'gdp': np.random.randint(1000, 20000, size=10),
        'population': np.random.randint(10000000, 1400000000, size=10),
        'life_expectancy': np.random.uniform(60, 85, size=10)
    })
    return data
```

```python
# Load and preprocess data
def preprocess_data(data):
    scaler = MinMaxScaler()
    data[['gdp', 'population', 'life_expectancy']] = scaler.fit_transform(data[['gdp', 'population', 'life_expectancy']])
    return data
```

```python
# Simple ML model
def predict_life_expectancy(data):
    X = data[['gdp', 'population']]
    y = data['life_expectancy']
    model = LinearRegression().fit(X, y)
    return model.predict(X)
```

```python
# Generate and preprocess data
data = generate_data()
data = preprocess_data(data)
```

```python
# Make predictions
data['predicted_life_expectancy'] = predict_life_expectancy(data)
```

```python
# Visualize results
plt.figure(figsize=(12, 6))
plt.scatter(data['gdp'], data['life_expectancy'], c='blue', label='Actual')
plt.scatter(data['gdp'], data['predicted_life_expectancy'], c='red', label='Predicted')
plt.xlabel('GDP (normalized)')
plt.ylabel('Life Expectancy (normalized)')
plt.title('Actual vs Predicted Life Expectancy')
plt.legend()
plt.show()

print(data)
```
